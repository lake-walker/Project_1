```python
# Begining with the transfer data from the last 5 years to gain our list of transfered players
# Import csv file for the Premier League for the last 5 years
epl_2016 = pd.read_csv('2016/english_premier_league.csv')
epl_2017 = pd.read_csv('2017/english_premier_league.csv')
epl_2018 = pd.read_csv('2018/english_premier_league.csv')
epl_2019 = pd.read_csv('2019/english_premier_league.csv')
epl_2020 = pd.read_csv('2020/english_premier_league.csv')
# Combine the new DataFrames into one
epl_frames = [epl_2016, epl_2017, epl_2018, epl_2019, epl_2020]
epl_total = pd.concat(epl_frames)
epl_total

# Clean the data to remove all transfers that were either free or with a loan association
# created a new columns with booleans telling me if a players 'fee' contained loan in it anywhere
epl_total['loaned?'] = epl_total['fee'].str.contains('loan', case=False)
epl_clean_data = epl_total[epl_total['loaned?'] == False]

# Remove the remaining non-values
epl_clean_data = epl_clean_data[(epl_clean_data['fee'] != '-') & (epl_clean_data['fee'] != '?') & (epl_clean_data['fee'] != 'Free transfer')]

# some of the fees are in thousands while others are in millions. For the sake of time I am only going to use those above 1 million pounds
epl_clean_data['cheap?'] = epl_clean_data['fee'].str.contains('Th', case=False)
epl_clean_data = epl_clean_data[epl_clean_data['cheap?'] == False]

# Clean it to only inbound transfers or 'purchases'
# Creating a sellers DataFrame aswell
epl_seller_data = epl_clean_data[epl_clean_data['transfer_movement'] == 'out']
epl_clean_data = epl_clean_data[epl_clean_data['transfer_movement'] == 'in']

# remove unneccessary columns
epl_clean_data = epl_clean_data.drop(columns={'loaned?','cheap?'})
# epl_clean_data
```
